```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, r2_score, accuracy_score
from sklearn.preprocessing import StandardScaler, LabelEncoder, MinMaxScaler
from sklearn.svm import SVC
```

```python
df = pd.read_csv('alevel_categorical.csv')
df.head()
df2 = df.copy()
```

```python
df = df.fillna(0)
df.head()
```

```python
df['Establishment'].unique()
```

```python
# le = LabelEncoder()

df['Establishment'] = LabelEncoder().fit_transform(df['Establishment'])
df.head(10)
```

```python
rep = {
    'A':5, 'B':4, 'C':3, 'D':2, 'E':1, 'F':0, 'U':0, 'X':0, 'S':0, 'O':0
}
df = df.replace(rep)
df.head()
```

```python
# df.info()
```

```python
# df = df.astype('float')
st_scaler = StandardScaler()
min_max = MinMaxScaler()
X = min_max.fit_transform(df.drop('Establishment', axis=1))
y = df.Establishment
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
```

```python
# Support Vector Classifier
svc = SVC()
svc.fit(X_train, y_train)
svc_pred = svc.predict(X_test)
accuracy_score(svc_pred, y_test)
```

```python
# Random Forest Classifier
random_model = RandomForestClassifier(n_estimators=258)
random_model.fit(X_train, y_train)
y_pred = random_model.predict(X_test)
accuracy_score(y_pred, y_test)
```

```python
# Decision Tree Classifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)
_pred = dtree.predict(X_test)
accuracy_score(_pred, y_test)
```

```python
acc = []
for i in range(255, 260):
    random_model = RandomForestClassifier(n_estimators=i)
    random_model.fit(X_train, y_train)
    y_pred = random_model.predict(X_test)  
    acc.append(accuracy_score(y_pred, y_test))
```

```python
import matplotlib.pyplot as plt
plt.plot(range(255,260),acc)
```
